```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn import preprocessing
```

```python
df = pd.read_csv('dataset.csv')
```

```python
df.head()
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
```

```python
df_train = df[df.year != 10]
```

```python
df_train.tail()
```

```python
df_test = df[df.year == 10]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model_new = Sequential()
x1 = df_train.iloc[:, 1:8].to_numpy()
y1= df_train.iloc[:,8:13].to_numpy()
# x1 = df_train[:, :8]
# y1= df_train[:,8:13]
print(y1.shape[1])
n_inputs, n_outputs = x1.shape[1], y1.shape[1]
model_new.add(Dense(20, activation = "relu", input_dim = n_inputs))
model_new.add(Dense(10, activation = "relu"))
# model_new.add(Dense(5, activation = "relu"))
model_new.add(Dense(n_outputs))
model_new.compile(optimizer = "adam", loss = "mse", metrics =[tf.keras.metrics.RootMeanSquaredError()])
model_new.fit(x1,y1, epochs = 1000, batch_size = 128)
```

```python
x2 = df_test.iloc[:,1:8]
y2 = df_test.iloc[:, 8:13]


result1 = model_new.evaluate(x2, y2)
```
